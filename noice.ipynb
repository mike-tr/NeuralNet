```python
import numpy as np
import matplotlib.pyplot as plt


noise_val = 45
noise_strength = 15
size = 480
max_val = 255

x = (np.random.rand(size) * noise_strength + noise_val )
x : np.ndarray = x.astype(int)
# x.shape
plt.bar(range(0,size), x)
plt.ylim(0, max_val)
#plt.bar(x, height=250)
```

```python
from numpy import random 
signals :np.ndarray = (np.random.rand(50) * (size - 20) + 10).astype(int)
print(signals)
x_copy = x.copy();
for i in signals:
    x_copy[i] = (random.random() * 0.5 + 0.5) * max_val
    for j in range(1, 10):
        v = random.random() * 0.25 + 0.75;
        x_copy[i - j] = x_copy[i-j+1] * v
        if(x_copy[i-j] < noise_strength):
            x_copy[i-j] = noise_strength
            break
plt.bar(range(0, size), x_copy)
plt.ylim(0, max_val)
```

```python
histo = np.zeros(max_val).astype(int)
for i in range(0, size):
    histo[x_copy[i]] += 1
#histo.sum()
histo.sum()
```

```python
hcumsum = histo.copy();
for i in range(1, max_val):
    hcumsum[i] += hcumsum[i-1]
hcumsum
```

```python
noise_percent = 0.1
calc_noise_level = 0
for i in range(1, max_val):
    if(hcumsum[i] > size * noise_percent):
        calc_noise_level = i
        break
calc_noise_level
```

```python
ypoints = np.ones(size) * calc_noise_level

plt.bar(range(0, size), x_copy)
plt.plot(ypoints, linestyle = 'dotted', color = 'red')
plt.ylim(0, max_val)
plt.show()
```

```python
x_copy_sin = x.copy();

i = int(size / 2)
r = 240
m = 100 - noise_val
x_copy_sin[i] = (random.random() * 0.5 + 0.5) * max_val
for j in range(1, r):
    d = j / r;
    d *= d;
    v = random.random() * 0.1;
    #print(m * (1- d))
    x_copy_sin[i+j] = m * (1 - d) * (1 - v) + noise_val
    x_copy_sin[i-j] = m * (1 - d) * (1 - v) + noise_val
plt.bar(range(0, size), x_copy_sin)
plt.ylim(0, max_val)
```

```python
histo = np.zeros(max_val).astype(int)
for i in range(0, size):
    histo[x_copy[i]] += 1

hcumsum = histo.copy();
for i in range(1, max_val):
    hcumsum[i] += hcumsum[i-1]

noise_percent = 0.1
calc_noise_level = 0
for i in range(1, max_val):
    if(hcumsum[i] > size * noise_percent):
        calc_noise_level = i
        break

ypoints = np.ones(size) * calc_noise_level

plt.bar(range(0, size), x_copy_sin)
plt.plot(ypoints, linestyle = 'dotted', color = 'red')
plt.ylim(0, max_val)
plt.show()
```
